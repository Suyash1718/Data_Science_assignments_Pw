```python
# 1. Reverse a string
s = "hello"
print(s[::-1])

# 2. Count the number of vowels in a string
s = "education"
vowels = 'aeiouAEIOU'
count = sum(1 for c in s if c in vowels)
print(count)

# 3. Check if a string is a palindrome
s = "madam"
print(s == s[::-1])

# 4. Check if two strings are anagrams
from collections import Counter
s1 = "listen"
s2 = "silent"
print(Counter(s1) == Counter(s2))

# 5. Find all occurrences of a substring
text = "the rain in spain falls mainly in the plain"
sub = "in"
positions = [i for i in range(len(text)) if text.startswith(sub, i)]
print(positions)

# 6. Basic string compression using counts of repeated characters
s = "aabcccccaaa"
compressed = ""
i = 0
while i < len(s):
    count = 1
    while i + 1 < len(s) and s[i] == s[i+1]:
        i += 1
        count += 1
    compressed += s[i] + str(count)
    i += 1
print(compressed)

# 7. Check if a string has all unique characters
s = "abcdefg"
print(len(s) == len(set(s)))

# 8. Convert string to uppercase and lowercase
s = "Python"
print(s.upper(), s.lower())

# 9. Count the number of words in a string
s = "This is a test string"
print(len(s.split()))

# 10. Concatenate two strings without using +
s1 = "Hello"
s2 = "World"
print("".join([s1, s2]))

# 11. Remove all occurrences of a specific element from a list
lst = [1, 2, 3, 2, 4, 2]
elem = 2
print([x for x in lst if x != elem])

# 12. Find second largest number in list
lst = [10, 20, 4, 45, 99]
unique = list(set(lst))
unique.sort()
print(unique[-2])

# 13. Count occurrences of each element in a list
lst = [1, 2, 2, 3, 1, 4, 2]
freq = {}
for i in lst:
    freq[i] = freq.get(i, 0) + 1
print(freq)

# 14. Reverse a list in-place without using reverse()
lst = [1, 2, 3, 4]
n = len(lst)
for i in range(n//2):
    lst[i], lst[n-i-1] = lst[n-i-1], lst[i]
print(lst)

# 15. Remove duplicates from list while preserving order
lst = [1, 2, 2, 3, 1, 4]
seen = set()
res = []
for x in lst:
    if x not in seen:
        res.append(x)
        seen.add(x)
print(res)

# 16. Check if list is sorted
lst = [1, 2, 3, 4]
print(lst == sorted(lst) or lst == sorted(lst, reverse=True))

# 17. Merge two sorted lists
a = [1, 3, 5]
b = [2, 4, 6]
merged = sorted(a + b)
print(merged)

# 18. Intersection of two lists
a = [1, 2, 3]
b = [2, 3, 4]
print(list(set(a) & set(b)))

# 19. Union of two lists without duplicates
a = [1, 2, 3]
b = [2, 3, 4]
print(list(set(a) | set(b)))

# 20. Shuffle a list without built-in shuffle
import random
lst = [1, 2, 3, 4, 5]
for i in range(len(lst)):
    j = random.randint(0, len(lst)-1)
    lst[i], lst[j] = lst[j], lst[i]
print(lst)

# 21. Common elements in two tuples
t1 = (1, 2, 3, 4)
t2 = (3, 4, 5, 6)
print(tuple(set(t1) & set(t2)))

# 22. Intersection of two sets entered by user
a = set(map(int, input("Enter first set: ").split(',')))
b = set(map(int, input("Enter second set: ").split(',')))
print(a & b)

# 23. Concatenate two tuples
def concat_tuples(t1, t2):
    return t1 + t2
print(concat_tuples((1, 2), (3, 4)))

# 24. Elements in first set but not in second
a = set(input("Set A: ").split())
b = set(input("Set B: ").split())
print(a - b)

# 25. Slice a tuple using indices
def slice_tuple(t, start, end):
    return t[start:end]
print(slice_tuple((1, 2, 3, 4, 5), 1, 4))

# 26. Union of two sets entered by user
a = set(input("Set A: ").split())
b = set(input("Set B: ").split())
print(a | b)

# 27. Max and Min from tuple using unpacking
t = (10, 20, 5, 30)
max_val, min_val = max(t), min(t)
print(max_val, min_val)

# 28. Union, intersection, difference of sets
a = {1, 2, 3}
b = {2, 3, 4}
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference:", a - b)

# 29. Count occurrences in tuple
def count_element(t, elem):
    return t.count(elem)
print(count_element((1, 2, 2, 3), 2))

# 30. Symmetric difference of sets
a = set(input("Set A: ").split())
b = set(input("Set B: ").split())
print(a ^ b)

# 31. Word frequency in list
words = ["apple", "banana", "apple", "orange"]
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)

# 32. Merge two dictionaries, add values of common keys
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
merged = d1.copy()
for k, v in d2.items():
    merged[k] = merged.get(k, 0) + v
print(merged)

# 33. Access nested dictionary value with list of keys
def get_nested(d, keys):
    for key in keys:
        d = d.get(key)
        if d is None:
            return None
    return d

data = {"a": {"b": {"c": 42}}}
print(get_nested(data, ["a", "b", "c"]))

# 34. Sort dictionary by values
d = {"a": 3, "b": 1, "c": 2}
sorted_d = dict(sorted(d.items(), key=lambda x: x[1]))
print(sorted_d)

# 35. Invert dictionary and group keys with same values
d = {'a': 1, 'b': 2, 'c': 1}
inverted = {}
for k, v in d.items():
    inverted.setdefault(v, []).append(k)
print(inverted)
```
